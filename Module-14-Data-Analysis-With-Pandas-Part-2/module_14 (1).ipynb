```python
# Custom Built Function
def grading_system(marks):
    if marks >=260:
        return 'A+'
    elif marks >=250:
        return 'A'
    else:
        return 'A-'


df["Grade"] = df['Total Marks'].apply(grading_system)
df
        
```

```python
# https://leetcode.com/problems/count-salary-categories/description/?envType=study-plan-v2&envId=30-days-of-pandas&lang=pythondata
# (df["Total Marks"] > 250).sum()
import pandas as pd

def count_salary_categories(accounts: pd.DataFrame) -> pd.DataFrame:
    low_salary = (accounts['income'] < 20000).sum()
    average_salary = ((accounts['income'] >= 20000)& (accounts['income'] <= 50000)).sum()
    high_salary = (accounts['income'] > 50000).sum()

    data = [('Low Salary',low_salary),('Average Salary',average_salary),("High Salary",high_salary)]

    df = pd.DataFrame(data,columns=['category','accounts_count'])

    return df
```

```python
df = pd.DataFrame({
    'A': [1, 2, np.nan, 4, 5],
    'B': [5, np.nan, np.nan, None, 10],
    'C': ['a', 'b', 'c', None, 'e']
})

df
```

```python
df["A"] = df["A"].fillna(df["A"].mean())
df["B"] = df["B"].ffill()
df["C"] = df["C"].apply(lambda x:"Unknown" if pd.isna(x) else x)
df
```

```python

print(df)

df_filled = df.fillna({'A':0,'B':99,'C':"Unknown"})
print(df_filled) 

df_filled1 = df.copy()
df_filled1['B'] = df_filled1["B"].ffill()
print(df_filled1) 

def f_missing(df):
    for col in df.columns:
        if col == "B":
            continue
        if pd.api.types.is_numeric_dtype(df[col]): 
            df[col] = df[col].fillna(df[col].mean())
        else:
            df[col] = df[col].fillna("Unknown")
    return df
df = f_missing(df)
print(df)
```

```python
# https://www.codechef.com/learn/course/pandas/LPDP06/problems/LPDPR28?tab=solution

import pandas as pd
import numpy as np

df = pd.DataFrame({
    'A': [1, 2, np.nan, 4, 5],
    'B': [5, np.nan, np.nan, None, 10],
    'C': ['a', 'b', 'c', None, 'e']
})
                   
print(df)  

# Drop rows where column A has missing values
df_clean1 = df.dropna(subset=["A"])
print(df_clean1)

# Drop rows with less than 2 non-null values
df_clean2 = df.dropna(thresh=2)
print(df_clean2)


# Drop rows where both Column A and Column B has missing values
df_clean3 = df.dropna(subset=['A','B'],how='all')
print(df_clean3)

```

```python
# https://www.codechef.com/learn/course/pandas/LPD10/problems/LPDPR46
import pandas as pd
import numpy as np

# Create a sample DataFrame
df = pd.DataFrame({
    'Category': ['A', 'B', 'A', 'B', 'A', 'B'],
    'Value1': [10, 20, 30, 40, 50, 60],
    'Value2': [100, 200, 300, 400, 500, 600]
})

# Basic groupby and aggregation
print("Mean of Value1 for each Category:")
print(df.groupby('Category')['Value1'].mean())

print("Max of Value2 for each Category:")
print(df.groupby('Category')['Value2'].max())

print("Sum of Value1 for each Category:")
print(df.groupby('Category')['Value1'].sum())

```
