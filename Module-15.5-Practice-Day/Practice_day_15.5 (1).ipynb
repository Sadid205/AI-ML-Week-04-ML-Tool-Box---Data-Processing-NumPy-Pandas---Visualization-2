```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv("Practice_day_15.5.csv",on_bad_lines='skip')
df
```

```python
# 1. Add New Column and Save File
# You have a DataFrame with column Price.
#  Create a new column Discounted_Price with (10% discounted price).
#  Save the modified DataFrame as discounted_products.csv.

df["Discounted_Price"] = (df["Price"] -  df["Price"] * 0.10)
df.to_csv("discounted_products.csv")
```

```python
# 2. Check Duplicates and Remove Them
# Column: EmployeeID
#  Tasks:
# Count duplicate entries


# Remove duplicates


# Display cleaned DataFrame
df.value_counts(df["EmployeeID"])
df.drop_duplicates(subset="EmployeeID",keep="first")
df
```

```python
# 3. Handle Missing Values
# Columns: Age, Salary
#  Tasks:
# Count missing values


# Fill missing Age with mean age


# Drop rows where Salary is missing

df = pd.read_csv("Practice_day_15.5.csv",on_bad_lines="skip")
df
print(df[["Age","Salary"]])
df[["Age","Salary"]].isna().sum()
df["Age"] = df["Age"].apply(lambda x: df["Age"].mean() if pd.isna(x) else x )
df
df.dropna(subset=["Salary"])
```

```python
# 4. Apply Statistical Functions
# Columns: Marks1, Marks2, Marks3
#  Tasks:
# Calculate mean, max for each subject


# Compute total marks per student


# Find standard deviation of total marks

df["Mean"] = df[["Marks1","Marks2","Marks3"]].mean(axis=1)
df["Max"] = df[["Marks1","Marks2","Marks3"]].max(axis=1)
df["Total Marks"] = df[["Marks1","Marks2","Marks3"]].sum(axis=1)
print("Standered deviation of Total Marks",df["Total Marks"].std())
```

```python
# 5. Histogram, Bar, Scatter, Pie
# Create:
# Histogram of Age
# Scatter plot (Age vs Salary)
# Scatter plot ( Price vs Discounted Price) 


plt.hist(df["Age"],color="green",edgecolor="black")
plt.show()
```

```python
plt.scatter(df["Age"],df["Salary"])
plt.show()
```

```python
plt.xlabel("Price",color="red")
plt.ylabel("Discounted Price",color="green")
plt.grid()
plt.scatter(df["Price"],df["Discounted_Price"])
plt.show()
```
