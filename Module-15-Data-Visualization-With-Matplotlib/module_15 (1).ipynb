```python
# Scatter Plot
students = pd.read_csv("student_IQdata.csv")
students
```

```python
# Bar Chart
data = {
    'Student': [ 'B', 'C', 'D', 'E', 'F', 'G', 'H'],
    'Gender': [ 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male'],
    'StudyHours': [ 2, 5, 3, 7, 4, 8, 1],
    'Attendance': [ 60, 88, 70, 95, 80, 98, 55],
    'Grade': [ 'Fail', 'Pass', 'Fail', 'Pass', 'Pass', 'Pass', 'Fail']
}

df = pd.DataFrame(data)
df
```

```python
# Gender wise catagories
gender_group = df.groupby('Gender').size()
gender_group.index
gender_group.values
```

```python
plt.xlabel("Gender")
plt.ylabel("Count")
plt.title("Bar Chart of Gender among students")
plt.bar(gender_group.index
,gender_group.values
,color=['orange','red'],edgecolor='black')
```

```python
# Bar Chart on real data
df = pd.read_csv("student_data.csv")
```

```python
status = df.groupby('CompletionStatus').size()
plt.xlabel("Completion Status")
plt.ylabel("Count")
plt.title("Bar Chart of course completion")
plt.bar(status.index
,status.values
,edgecolor='black',color=["green","orange","red"])
status.index
plt.show()
```

```python
status = df.groupby('Location').size()
plt.xlabel("Location")
plt.ylabel("Count")
plt.title("Bar Chart of students location")
plt.bar(status.index
,status.values
,edgecolor='black',color=["green","orange","red"])
status.index
plt.show()
```

```python
# Pie Chart
status = df.groupby('CompletionStatus').size() 
explode = (0.1,0.1,0.1) 
plt.pie(status,labels=status.index,shadow=True,explode=explode,autopct='%1.1f%%')
plt.show()
```
